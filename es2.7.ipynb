```python
from pylab import *
```

```python
gvalue = lambda x,n : sin(x)/x**n
```

```python
x = linspace(-5,5,5000)
from matplotlib.pyplot import figure
fig = figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
ylim(-50,150)

for i in range(1,4):
    y = gvalue(x, i)
    plot(x,y, figure = fig, label = "sin(x)/x**"+str(i))
    
legend()
show()
```
