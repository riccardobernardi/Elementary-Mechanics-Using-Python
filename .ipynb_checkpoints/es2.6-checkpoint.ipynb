```python
from pylab import *
```

```python
fvalue = lambda x, n: x**(-n)
```

```python
x = linspace(1,11,1000)

for i in range(1,12):
    # print(i/2)
    y = fvalue(x,i/2)
    subplot(2,3,i%6+1)
    plot(x,y)
show()
```

```python
x = linspace(1,11,1000)

for i in range(1,12):
    # print(i/2)
    y = fvalue(x,i/2)
    subplot(2,3,1)
    plot(x,y)
show()
```

```python
from matplotlib.pyplot import figure
```

```python
x = linspace(-1,1,50)

for i in range(1,4):
    # print(i/2)
    y = fvalue(x,i)
    # subplot(2,3,1)
    plot(x,y,figure =  figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k'))
    show()
```

```python
x = linspace(-1,1,50)
fig = figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
ylim(-1000,1000)

for i in range(1,4):
    # print(i/2)
    y = fvalue(x,i)
    # subplot(2,3,1)
    plot(x,y,figure = fig, label = "x**(-"+str(i)+")")
legend()
show()
```
