```python
from pylab import *
```

```python
def normal(x, mu, si):
    return (1/sqrt(2*pi*si**2))*exp(-(x-mu)**2/(2*si**2))
```

```python
x = linspace(-5,5,1000)
y = normal(x, 0, 1)
plot(x,y)
show()
```

```python
x = linspace(-5,5,1000)
y = normal(x, 0, 2)
plot(x,y)
show()
```

```python
subplot(2,3,1)
x= linspace(-5,5,1000)
y = normal(x, 0, 0.01)
plot(x,y)
show()

subplot(2,3,2)
x = linspace(-5,5,1000)
y = normal(x, 0, 1)
plot(x,y)
show()

subplot(2,3,3)
x = linspace(-5,5,1000)
y = normal(x, 0, 2)
plot(x,y)
show()
```
