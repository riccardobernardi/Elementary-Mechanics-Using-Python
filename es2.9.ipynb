```coconut
from pylab import *
```

```coconut
acc = (v,x,k,c) -> -k*x -c*v
```

```coconut
def acceleration(k,c):
    dt = 0.01
    t = 100*dt
    n = 100
    xt = zeros(n, float)
    xt[0] = 0
    vt = zeros(n, float)
    vt[0] = 1
    at = zeros(n, float)
    for i in range(1,100):
        xt[i] = xt[i-1] + vt[i-1]*dt
        at[i] = acc(vt[i-1],xt[i-1], k, c)
        vt[i] = vt[i-1] + at[i-1]*dt
    title("space")
    plot(xt)
    show()
    title("velocity")
    plot(vt)
    show()
    title("acceleration")
    plot(at)
    show()
```

```coconut
acceleration(10,5)
```
