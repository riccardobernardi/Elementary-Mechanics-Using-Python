```coconut
# given x(i) and r as input return x(i+1)
# x(i+1) = r*x(i)*(1-x(i))

logistic = (x,r) -> r*x*(1-x)
```

```coconut
def logit(r):
    x = [0.5]
    for i in range(1,100):
        x += [logistic(x[i-1], r)]
    plot(x), show()
```

```coconut
logit(1.0)
```

```coconut
def logit(r):
    n=100
    x = zeros(n,float)
    x[0] = 0.5
    for i in range(n-1):
        x[i+1] = logistic(x[i], r)
    i = r_[0:n-1]; plot(i, x[:99]), show()
```

```coconut
logit(1.0)
```

```coconut
logit(2.0)
```
