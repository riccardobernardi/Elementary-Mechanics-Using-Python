```python
x0 = 0.0; x1 = 10.0; dx = 0.1
n = int(ceil((x1-x0)/dx) + 1)
x = zeros((n,1), float)
y = zeros((n,1), float)
for i in range(n):
    x[i] = x0 + i*dx
    y[i] = sin(x[i])
plot(x,y)
show()
```

```python
t0 = 0.0; t1 = 10.0; dt = 0.1
n = int(ceil((t1-t0)/dt) + 1)
t = zeros((n,1), float)
y = zeros((n,1), float)
t[0] = t0
y[0] = 100.0 - 4.9*t[0]**2
i = 0
while y[i]>0.0:
    i = i+1
    t[i] = t0 + i*dt
    y[i] = 100.0 - 4.9*t[i]**2
plot(t[0:i],y[0:i])
xlabel('t[s]')
ylabel('y[m]')
show()
```

```python
x = linspace(-5,5,1000)

normale = lambda x: exp(-x**2)

y = normale(x)

plot(x,y)
show()
```

```python
def derivative(f,x):
    h = 0.0001
    return lambda x: (f(x+h) - f(x-h))/(2*h)
```

```python
plot(x, derivative(normale, x)(x))
```
