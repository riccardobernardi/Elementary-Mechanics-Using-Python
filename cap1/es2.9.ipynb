```coconut
from pylab import *
```

```coconut
acc = (v,x,k,c) -> -k*x -c*v
```

```coconut
def acceleration(k,c):
    dt = 0.01
    t0 = 0
    n = 100
    xt = zeros(n, float)
    xt[t0] = 0
    vt = zeros(n, float)
    vt[t0] = 1
    t = zeros(n, float)
    at = zeros(n, float)
    for i in range(1,100):
        xt[i] = xt[i-1] + vt[i-1]*dt
        at[i] = acc(vt[i-1],xt[i-1], k, c)
        vt[i] = vt[i-1] + at[i-1]*dt
        t[i] = t[i-1] + dt
    title("space")
    plot(t,xt)
    show()
    title("velocity")
    plot(t,vt)
    show()
    title("acceleration")
    plot(t,at)
    show()
```

```coconut
acceleration(10,5)
```
